```python
def simulated_inning_team(transition_matrix):
    """
    Given a team transition matrix, this outputs the runs scored in an example inning, 
    as well as the states of the inning, starting with (0,'') and ending with either
    '0', '1', '2', or '3'.
    """
    possible_states=[(0,''), (0,'1'), (0,'2'), (0,'3'), (0, '12'), (0,'13'), (0,'23'), (0,'123'),
                        (1,''), (1,'1'), (1,'2'), (3,'3'), (1, '12'), (1,'13'), (1,'23'), (1,'123'),
                        (2,''), (2,'1'), (2,'2'), (2,'3'), (2, '12'), (2,'13'), (2,'23'),(2,'123'), 
                         '0', '1', '2', '3']
      
    current_state=np.random.choice(transition_matrix.index, size=1, p=transition_matrix.iloc[0,:])[0]
    all_states=[(0,''),current_state]
    while current_state in possible_states[0:23]:
        next_state=np.random.choice(transition_matrix.index, size=1, 
                                p=transition_matrix.ix[[current_state], :].values[0])[0]
        all_states+=[next_state]
        current_state=next_state
    runmatrix=run_matrix()
    runs=0
    for i in range(len(all_states)-1):
        runs+=runmatrix.ix[[all_states[i]], [all_states[i+1]]].values[0][0]
    return runs, all_states
```

```python
for i in range(100):
    print(simulated_inning_team(NYY_transition_matrix))
```

```python
def simulated_game_team(transition_matrix):
    """
    Given a team transition matrix, runs through 9 simulated innings. 
    """
    innings=[]
    for i in range(9):
        innings+=[simulated_inning_team(transition_matrix)[0]]
    return sum(innings)
```

```python
#Finding average runs per game for a season. Takes a bit to ru
runs=[]
for i in range(162):
    runs+=[simulated_game_team(NYY_transition_matrix)]
sum(runs)/162
```
