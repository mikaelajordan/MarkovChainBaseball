```python
import pandas as pd
import numpy as np
import os
from markov_functions import *
```

```python
player_types=pd.read_csv("../Baseball_Data/predicted_player_types.csv")
player_names=pd.read_csv("../Baseball_Data/player_id.csv")
player_names=pd.concat([player_names['bref_id'], player_names['retro_id']], axis=1)
player_types.rename(columns={'Unnamed: 0':'bref_id'}, inplace=True)
```

```python
player_types=pd.merge(player_types, player_names, on='bref_id',how='left')
```

```python
retro_type=pd.concat([player_types['retro_id'], player_types['Type']], axis=1)
retro_type.rename(columns={'retro_id':'resbatter'}, inplace=True)
```

```python
#collecting batting data
os.chdir("../Baseball_Data/Play_by_Play/")
batting=import_raw_batting_data()
os.chdir("../Optimization/")
```

```python
#adds player type to the batting data
batting_w_type=pd.merge(batting, retro_type, on='resbatter', how='left')
```

```python
#Makes a length 4 list where each element is the batting data for that type player
batting_by_type=[batting_w_type[batting_w_type['Type']==x] for x in range(0,4)]
```

```python
transition_matrices_by_type=[make_transition_matrix(i) for i in batting_by_type]
```

```python
player_types=player_types.set_index(player_types['ret'])
del player_types.index.name
player_types.drop('bref_id', axis=1, inplace=True)
```

```python
def combo_player_transition_combo(player_id):
    """
    Given a player id, returns a linear combination of the player type transition matrices.
    Since the function reads in the player_types.csv, must change working directory to Optimization. 
    To avoid rerunning the transition matrices calulations, run the 
    """
    mat=[]
    for i in range(4):
        mat+=[(player_types.ix[player_id,i]*transition_matrices_by_type[i])]
    
    return mat
```
