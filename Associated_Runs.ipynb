Here is the code for building an associated runs matrix. Each positon (i,j) in the matrix represents the runs associated with goint from position i to j. Not every position is possible, and therefore there are some positions that have nonsense, such as scoring 6 runs in one at bat. Every possible scenario has an associated run value that makes sense. 

```python
import pandas as pd
import numpy as np
```

```python
#Makes possible states and creates an empty 28 by 28 matrix, with rows and columns properly named
states=[(0,''), (0,'1'), (0,'2'), (0,'3'), (0, '12'), (0,'13'), (0,'23'), (0,'123'),
        (1,''), (1,'1'), (1,'2'), (3,'3'), (1, '12'), (1,'13'), (1,'23'), (1,'123'),
        (2,''), (2,'1'), (2,'2'), (2,'3'), (2, '12'), (2,'13'), (2,'23'),(2,'123'), '0', '1', '2', '3']
R1=pd.DataFrame(np.reshape(np.zeros(28*28), (28,28)))
R1.columns=states
R1.index=states
R1
```

```python
#Populates the matrix with associated runs. Some spots have numbers that make no sense, 
#but they only appear in impossible situations.
for i in states:
    if len(i)==2:
        i_out=i[0]
        i_runners=len(str(i[1]))
        for j in states:
            if len(j)==2:
                j_out=j[0]
                j_runners=len(str(j[1]))
                R1.ix[i,j]=(i_out+i_runners+1)-(j_out+j_runners)
            else:
                R1.ix[i,j]=int(j)
    else: 
        for j in states:
            R1.ix[i,j]=int(i)
R1[R1<0]=0
```
