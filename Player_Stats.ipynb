```python
import numpy as np
import pandas as pd
```

```python
#importing data from the Sean Lahman database. Super great and comprehensive data, but we don't need everything
player_table=pd.read_csv('./Batting.csv')
players_2016=player_table[(player_table.ix[:,'yearID']==2016)]
```

```python
players_2016=players_2016.set_index(players_2016['playerID'])
remove=['yearID', 'teamID', 'stint','lgID', 'G', 'playerID']
for i in remove:
    players_2016.drop(i, axis=1, inplace=True)
del players_2016.index.name
players_2016=players_2016[players_2016['AB']>50]
```

```python
player_names=pd.read_csv('./player_id.csv')
player_names=pd.concat([player_names['lahman_id'], player_names['mlb_name']], axis=1)
player_names=player_names.set_index(player_names['lahman_id'])
player_names.drop('lahman_id', axis=1, inplace=True)
del player_names.index.name
player_names
```

```python
players_2016=players_2016.join(player_names)
players_2016=players_2016[list(players_2016.columns.values)[-1:]+list(players_2016.columns.values)[:-1]]
```

```python

#Clusters the baseball players with 4 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(players_2016.iloc[:,2:])
y_kmeans = kmeans.predict(players_2016.iloc[:,2:])
```

```python
#Number of players in each class
players_2016['player_type']=y_kmeans
for i in range(0,4):
    temp=len(players_2016[players_2016['player_type']==i])
    print(temp)
```
