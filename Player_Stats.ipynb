```python
import numpy as np
import pandas as pd
```

```python
#importing data from the Sean Lahman database. Super great and comprehensive data, but we don't need everything
player_table=pd.read_csv('./baseballdatabank-2017.1/core/Batting.csv')
players_2016=player_table[(player_table.ix[:,'yearID']==2016)]
```

```python
players_2016=players_2016.set_index(players_2016['playerID'])
remove=['yearID', 'teamID', 'stint','lgID', 'G', 'playerID']
for i in remove:
    players_2016.drop(i, axis=1, inplace=True)
del players_2016.index.name
players_2016=players_2016[players_2016['AB']>50]
```

```python
player_names=pd.read_csv('./player_id.csv')
player_names=pd.concat([player_names['lahman_id'], player_names['mlb_name']], axis=1)
player_names
```

```python
for i in range(len(players_2016)):
    for j in range(len(player_names)):
        if players_2016.index[i]==player_names['lahman_id'][j]:
            players_2016['names'][i]=player_names['mlb_name'][j]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(players_2016.iloc[:,1:])
y_kmeans = kmeans.predict(players_2016.iloc[:,1:])
```

```python
players_2016['player_type']=y_kmeans
for i in range(1,4):
    temp=len(players_2016[players_2016['player_type']==i])
    print(temp)
```

```python
def avg_deviation_from_center(k, data_set):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_set)
    y_kmeans = kmeans.predict(data_set)
    data_set['class']=y_kmeans
    dist_list=[]
    for classifier in range(1,k+1):
        for row_num in range(len(data_set[data_set['class']==classifier])):
            dist=0
            for col_num in range(len(data_set[data_set['class']==classifier].iloc[row_num,:])):
                dist+=(data_set[data_set['class']==classifier].iloc[row_num,col_num]- kmeans.cluster_centers_[classifier][col_num-1])**2
            dist_list+=[dist**.5]
    return sum(dist_list)/len(dist_list)
```

```python

dists=[]
for i in range(2,10):
    dists+=[avg_deviation_from_center(i, players_2016.iloc[:,1:])]
    
dists
    
```
